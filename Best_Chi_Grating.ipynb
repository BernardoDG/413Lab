```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from math import pi

print(os.getcwd())
```

```python
df=pd.read_csv('new fit.csv')
df=df.dropna()
df['mlamb'] = df['mlamb'].astype(float, errors = 'raise')

x = pi/180*df.theta
y = df.mlamb
#err = df.error
th_i1 = -3.1375*pi/180
d1 = 16668.047

def f(x,th_i,d):
    return d*(np.sin(th_i)+np.sin(x-th_i))

#fit = []
chi = []
a = np.linspace(16668.047-1.3,16668.047+1.3,100)
for i in a:
    s = 0
    for j in range(0,20):
        error = i*np.cos(x[j]-th_i1)*pi/10800
        fit_1 = f(x[j],th_i1,i)
        s += ((y[j]-fit_1)/error)**2
    chi.append(s)

plt.figure(figsize=(8,8))
plt.plot(a,chi)
plt.xlabel('d (angstroms)')
plt.ylabel('chi square')
plt.grid()
plt.show()
print(min(chi))
```
